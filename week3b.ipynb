# Segmenting and Clustering Neighborhoods in Toronto #



#### importing libraries and web scraping ####

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source, 'lxml')
x=soup.table.text
new_string=''
y=x[34:]
for line in y.split('\n'):
        if line.strip():
                new_string += line + '\n'
t=new_string
print(t)
```

#### making dataframe ####

```python
column_names = ['Postcode', 'Borough', 'Neighborhood'] 

i=[]
t=t.rstrip()
i=t.split('\n')
w=''
for r in i:
    if r==w:
        i.remove(r)
df=pd.DataFrame(np.array(i).reshape(len(i)//3,3),columns=column_names)
df.head(10)
```

#### assigning neighborhood(not assigned) values to value of corresponding Borough ####

```python

df2=df[df.Borough!='Not assigned']
df2['Neighborhood'] = df2.apply(
    lambda row: row['Borough'] if row['Neighborhood'] == 'Not assigned' else row['Neighborhood'],
    axis=1)
df2.head()
```

#### creating mutliple value for neighborhood ####

```python
df_table4=df2.copy()
df_table4 = (df_table4.groupby(['Postcode','Borough'])['Neighborhood']
       .apply(lambda x: ','.join(set(x.dropna())))
       .reset_index())

df_table4 = pd.DataFrame(df_table4)
df_table4.head(10)
```
